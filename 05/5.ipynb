```python
with open("input_5.txt", "r") as f:
    data = [l for l in f.read().split("\n") if l]
data
```

```python
def get_seat_nb(seat, bottom_letter, top_letter, range_end, range_start=0):    
    current_letter = seat[0]
    
    row_range = (range_end - range_start + 1) // 2

    if current_letter == bottom_letter:
        range_end -= row_range
    elif current_letter == top_letter:
        range_start += row_range
    
    if range_start == range_end:
        return range_start
    
    return get_seat_nb(seat[1:], bottom_letter, top_letter, range_end, range_start)

def parse_seat(seat):
    row = get_seat_nb(seat[:7], "F", "B", 127)
    column = get_seat_nb(seat[7:], "L", "R", 7)
        
    return row, column

def get_seat_id(row, column):
    return row * 8 + column
```

```python
seat_ids = [get_seat_id(*parse_seat(seat)) for seat in data]
highest_seat_id = max(seat_ids)
highest_seat_id
```

```python
sorted_seats = sorted(seat_ids)
sorted_seats
```

```python
seat_distance = [(i, s - sorted_seats[i]) for i, s in enumerate(sorted_seats[1:])]
my_seat_idx = [i for i, dist in seat_distance if dist == 2][0]
my_seat_id = sorted_seats[my_seat_idx] + 1
my_seat_id
```

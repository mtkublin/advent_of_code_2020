```python
def get_distinct_answers(group_answers):
    return {ans for person_ans in group_answers for ans in person_ans}

def count_distinct_answers(group_answers):
    return len(get_distinct_answers(group_answers))

def count_all(answers):
    return sum(count_distinct_answers(group) for group in data)
```

```python
def get_distinct_answers(group_answers):
    result = set(group_answers[0])
    for next_ans in group_answers[1:]:
        result = result.intersection(set(next_ans))
    return result

def count_distinct_answers(group_answers):
    return len(get_distinct_answers(group_answers))

def count_all(answers):
    return sum(count_distinct_answers(group) for group in data)
```

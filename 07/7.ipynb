```python
with open("input_7.txt", "r") as f:
    data = [d for d in f.read().split("\n") if d]
data
len(data)
```

```python
def get_color_number(color):
    if color == " no other bags.":
        return None
    split_color = color.split(" ")
    color_name = " ".join(split_color[2:4])
    nb = split_color[1]
    return color_name, int(nb)

def get_stored_colors(bag):
    bag_color, stored_colors = bag.split(' bags contain')
    stored_colors = [get_color_number(color) for color in stored_colors.split(",")]

    return bag_color, dict([c for c in stored_colors if c is not None])

def parse_rules(rules):
    return dict([get_stored_colors(r) for r in rules])
```

```python
rule_dict = parse_rules(data)
len(rule_dict)
```

```python
import json
print(json.dumps(rule_dict, indent=4))
```

```python
def find_direct_color_containers(bags, searched_color="shiny gold"):
    return {color for color, bag_content in bags.items() if searched_color in bag_content}

def find_all_color_containers(bags, searched_colors=None):
    if searched_colors is None:
        searched_colors = {"shiny gold"}
        
    direct_containers = {cont for color in searched_colors for cont in find_direct_color_containers(bags, color)}
    if not direct_containers:
        return direct_containers
    
    return {
        *direct_containers,
        *find_all_color_containers(bags, searched_colors=direct_containers)
    }
```

```python
all_shiny_gold_containers = find_all_color_containers(rule_dict)
all_shiny_gold_containers
```

```python
def get_inner_bags_nb(bags, color='shiny gold'):    
    direct_bags = bags[color]
    
    bags_nb = 0
    for cl, count in direct_bags.items():
        bags_nb += count + count * get_inner_bags_nb(bags, cl)
    
    return bags_nb
```

```python
bag_nbs = get_inner_bags_nb(rule_dict)
```

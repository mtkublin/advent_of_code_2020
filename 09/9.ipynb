```python
def find_possible_components(sum_nb, initial_components):
    first_components = set(initial_components)
    second_components = {sum_nb - cmp for cmp in initial_components}
    
    possible_components = {cmp for cmp in first_components.intersection(second_components) if cmp * 2 != sum_nb}
        
    return possible_components
    
def check_if_valid(nb, possible_components):
    return bool(find_possible_components(nb, possible_components))

def find_first_invalid(numbers, preambule_size=25):
    for idx, nb in enumerate(numbers[preambule_size:]):
        components = numbers[idx: idx+preambule_size]
        if check_if_valid(nb, components):
            continue
        return nb
```

```python
def first_solution(sum_nb, components):
    for start_idx, cmp in enumerate(components):
        for end_idx in range(start_idx + 2, len(components)):
            current_set = set(components[start_idx: end_idx])
            if sum(current_set) == sum_nb:
                return current_set
            
def find_components_set_border_sum(sum_nb, components):
    comp_set = first_solution(sum_nb, components)
    return min(comp_set) + max(comp_set)
```

```python
def grow(start_idx, end_idx, sum_nb, components):
    while sum(components[start_idx:end_idx]) < sum_nb:
        end_idx += 1
    return start_idx, end_idx


def shift(start_idx, end_idx, sum_nb, components):
    start_idx += 1
    return start_idx, end_idx


def shrink(start_idx, end_idx, sum_nb, components):
    while sum(components[start_idx:end_idx]) > sum_nb:        
        end_idx -= 1
    return start_idx, end_idx

def change_gen():
    idx = 0
    while True:
        try:
            yield (grow, shift, shrink)[idx]
        except IndexError:
            idx = 0
        else:
            idx += 1

def crawl(sum_nb, components):
    start_idx, end_idx = 0, 2
    change_generator = change_gen()
    while sum(components[start_idx:end_idx]) != sum_nb:
        change = next(change_generator)
        start_idx, end_idx = change(start_idx, end_idx, sum_nb, components)
    return components[start_idx:end_idx]
```
